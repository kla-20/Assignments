<h2>Solve Reddy Mikks by Graphical Method</h2> 
<h5>To construct objective function, note that the company wants to maximize (i.e.
,increase much as possible) the total daily profit of both paints. Given that the profit
per ton of exterior and interior paints are 5 and 4 (thousand) dollars, respectively, it
follows that

Total profit from exterior paint=5x

Total profit from exterior paint=4y

Letting z represent the total daily profit, the objective of the company is</h5>




    Maximize Z= 5x+4y
    subject to:
        1.6x + 4y <= 24
        2.x + 2y <= 6
        3.x>0
        4.y>0
        5.y<=2
        6.y-x<=1

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sympy import *
```

```python
def eqn1(x): # for eqn 1
    """
    Summary: returns y value for eqn for 6x + 4y = 24
    Args:
        x - variable , type-float
    Author: Lalith  
    """
    return (24-6*x)/4

def eqn2(x): # for eqn 2
    """
    Summary: returns y value for eqn for x + 2y = 6
    Args:
        x - variable , type-float
    Author: Lalith  
    """  
    return (6-x)/2

def xax(): # for eqns 3,4
    """
    Summary: returns 0 since x=0, y=0 for x-axis
    Args:
        None
    Author: Lalith  
    """
    return 0

def eqn6(x):
    """
    Summary: returns y value for eqn for y - x - 1 = 0
    Args:
        x - variable , type-float
    Author: Lalith  
    """
    return x+1

x1 = np.linspace(-2,10,100)
y1 = []
for i in range(len(x1)): # for eqn 1
    y1.append([eqn1(x1[i])])

x2 = np.linspace(-2,10,100)
y2=[]
for i in range(len(x2)): # for eqn 2
    y2.append([eqn2(x2[i])])
    
x3 = np.linspace(-2,10,100)
y3=[]
for i in range(len(x3)): # for eqn 3,4
    y3.append([xax()])
    
x4 = np.linspace(0,10,100)
y4=[]
for i in range(len(x4)): # for eqn 5
    y4.append([2])
    
x5 = np.linspace(-2,10,100)
y5=[]
for i in range(len(x5)): # for eqn 6
    y5.append([eqn6(x5[i])])    
    
x, y = symbols('x y')

eq1 = Eq((6*x + 4*y -24),0)
eq2 = Eq((x+ 2*y -6),0)
eq3 = Eq((x),0)
eq4 = Eq((y),0)
eq5 = Eq((y-2),0)
eq6 = Eq((y-x-1),0)
# We will solve the equations now

#Solving equations 1 and 2 we get soln 1
sol1 = solve((eq1, eq2),(x, y))
dummy = []
dummy.append(sol1[x])
dummy.append(sol1[y])
sol1 = dummy


#Solving equations 1 and 3 we get soln 2
sol2 = solve((eq1, eq3),(x, y))
dummy = []
dummy.append(sol2[x])
dummy.append(sol2[y])
sol2 = dummy


#Solving equations 1 and 4 we get soln 3
sol3 = solve((eq1, eq4),(x, y))
dummy = []
dummy.append(sol3[x])
dummy.append(sol3[y])
sol3 = dummy

#Solving equations 1 and 5 we get soln 4
sol4 = solve((eq1, eq5),(x, y))
dummy = []
dummy.append(sol4[x])
dummy.append(sol4[y])
sol4 = dummy

#Solving equations 1 and 6 we get soln 5
sol5 = solve((eq1, eq6),(x, y))
dummy = []
dummy.append(sol5[x])
dummy.append(sol5[y])
sol5 = dummy


#Solving equations 2 and 3 we get soln 6
sol6 = solve((eq2, eq3),(x, y))
dummy = []
dummy.append(sol6[x])
dummy.append(sol6[y])
sol6 = dummy


#Solving equations 2 and 4 we get soln 7
sol7 = solve((eq2, eq4),(x, y))
dummy = []
dummy.append(sol7[x])
dummy.append(sol7[y])
sol7 = dummy


#Solving equations 2 and 5 we get soln 8
sol8 = solve((eq2, eq5),(x, y))
dummy = []
dummy.append(sol8[x])
dummy.append(sol8[y])
sol8 = dummy

#Solving equations 2 and 6 we get soln 9
sol9 = solve((eq2, eq6),(x, y))
dummy = []
dummy.append(sol9[x])
dummy.append(sol9[y])
sol9 = dummy

#Solving equations 3 and 4 we get soln 10
sol10 = solve((eq3, eq4),(x, y))
dummy = []
dummy.append(sol10[x])
dummy.append(sol10[y])
sol10 = dummy

#Solving equations 3 and 5 we get soln 11
sol11 = solve((eq3, eq5),(x, y))
dummy = []
dummy.append(sol11[x])
dummy.append(sol11[y])
sol11 = dummy

#Solving equations 3 and 6 we get soln 12
sol12 = solve((eq3, eq6),(x, y))
dummy = []
dummy.append(sol12[x])
dummy.append(sol12[y])
sol12 = dummy

#Solving equations 4 and 5 we get soln 13 but it cannot exist
"""
sol13 = solve((eq4, eq5),(x, y))
dummy = []
print(sol13)
dummy.append(sol13[x])
dummy.append(sol13[y])
sol13 = dummy
"""

#Solving equations 4 and 6 we get soln 14
sol14 = solve((eq4, eq6),(x, y))
dummy = []
dummy.append(sol14[x])
dummy.append(sol14[y])
sol14 = dummy

#Solving equations 5 and 6 we get soln 15
sol15 = solve((eq5, eq6),(x, y))
dummy = []
dummy.append(sol15[x])
dummy.append(sol15[y])
sol15 = dummy


#plot the equations    
plt.figure(figsize=(15,15))
plt.plot(x1,y1,color='red',label='6x+4y=24')
plt.plot(x2,y2,color='darkgreen',label='x+2y=6')
plt.plot(x3,y3,color='black',label='x=0')
plt.plot(y3,x3,color='black',label='y=0')
plt.plot(x4,y4,color='blue',label='y=2')
plt.plot(x5,y5,color='purple',label='y-x=1')

#Scatter all the solutions
plt.scatter(sol1[0],sol1[1],c='red',marker='X',s=200,label='Feasible_Point')
plt.scatter(sol2[0],sol2[1],c='blue',marker='o',s=100)
plt.scatter(sol3[0],sol3[1],c='red',marker='X',s=200)
plt.scatter(sol4[0],sol4[1],c='green',marker='o',s=150)
plt.scatter(sol5[0],sol5[1],c='black',marker='o',s=150)
plt.scatter(sol6[0],sol6[1],c='black',marker='o',s=150)
plt.scatter(sol7[0],sol7[1],c='black',marker='o',s=150)
plt.scatter(sol8[0],sol8[1],c='red',marker='X',s=200)
plt.scatter(sol9[0],sol9[1],c='black',marker='o',s=150)
plt.scatter(sol10[0],sol10[1],c='red',marker='X',s=200)
plt.scatter(sol11[0],sol11[1],c='blue',marker='o',s=150)
plt.scatter(sol12[0],sol12[1],c='red',marker='X',s=200)
plt.scatter(sol14[0],sol14[1],c='black',marker='o',s=150)
plt.scatter(sol15[0],sol15[1],c='red',marker='X',s=200)
plt.title('Graphical Method for Reddy Mikks Problem',fontsize=25)

# Feasible points
ansx = [sol1[0],sol3[0],sol8[0],sol10[0],sol12[0],sol15[0]]
ansy = [sol1[1],sol3[1],sol8[1],sol10[1],sol12[1],sol15[1]]

#Maximize Z = 5x + 4y
z = []
for i in range(len(ansx)):
    z.append(5*ansx[i] + 4*ansy[i])
for i in range(1,len(z)):
    initial = z[0]
    if(z[i]>initial):
        initial = z[i]

xo = initial
for i in range(len(z)):
    if(xo==z[i]):
        index = i
        break
        
plt.scatter(ansx[index],ansy[index],c='deeppink',marker='D',s=280,label='Final_Solution')   
plt.scatter(ansx[index],ansy[index],c='yellow',marker='+',s=200) 
plt.legend(fontsize='20',loc=1)
#plt.fill(ansx,ansy,0,facecolor='mediumpurple',color='blue',alpha=.5)

k = np.linspace(-3,10,1000)

x,y = np.meshgrid(k,k)
plt.imshow(((6*x+4*y<=24) & (x+2*y<=6) & (-x+y<=1) & (y<=2)) & (x>=0) & (y>=0).astype(int) , 
                extent=(x.min(),x.max(),y.min(),y.max()),origin="lower", cmap="GnBu", alpha = 0.39);
plt.xlim(-3,10)
plt.ylim(-3,10)
plt.show()

print(f"The Function Z= 5x + 4y is maximum at x={ansx[index]},y={ansy[index]} with maximum value: {z[index]}")
```

<h3>Answer: At 3 Tonns Exterior Paints and 3/2 Tonns Interior Paints , The company will make maximum profit with 21000$</h3>
