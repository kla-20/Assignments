```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sympy import *
```

```python
#1  Binomial Distribution

def bin_prob(n,p):
    """
    Summary: This function will take number of independent events and probability and creates a list based on n and returns list based on Binomial Distribution
    Args: n - number of independent bernoulli events , type- int
          p - probability of success
    Author: Lalith      
    """
    x = np.arange(n)
    f = lambda x: math.comb(n,x)*(p**x)*((1-p)**(n-x))
    y = []
    for i in range(0,n):
        y.append(f(x[i]))
    return y    
    
n = int(input("Enter n:"))
p = float(input("Enter p(0<=p<=1):"))

x= np.arange(n)

plt.bar(x,bin_prob(n,p),width=.25,color='#88c999')
plt.xlabel('x',fontsize=18)
plt.ylabel('Bin(n,p)',fontsize=18)
plt.title('Binomial Distribution',fontsize=20)
plt.plot(x,bin_prob(n,p))
plt.show()
```

```python
# 2.Normal Distribution

def normal(mean,sd):
    """
    Summary: This function will take mean and sd and creates a x - list and corresponding y-list based on normal distribution and returns y-list
    Args: mean - mean of the distribution , type - float
          sd - standard deviation of the distribution, type - float
    Author: Lalith      
    """
    x = np.linspace(mean-4*sd,mean+4*sd,100)
    f = lambda x: 1/(sd*np.sqrt(2*np.pi))*np.exp(-0.5*((x-mean/sd)**2))
    y = []
    for i in range(len(x)):
        y.append(f(x[i]))
    return y

mean = float(input("Enter mean:"))
sd = float(input("Enter standard deviation:"))

x = np.linspace(mean-4*sd,mean+4*sd,100)


plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)
plt.title('Normal Distribution',fontsize=20)
plt.plot(x,normal(mean,sd),color = 'red')
plt.scatter(x,normal(mean,sd),color = '#88c999')
plt.show()
```

```python
# 3.Poisson Distribution


def poisson(lam,n):
    """
    Summary: This function will take lambda and n and creates a x - list and a corresponding y-list based on poisson distribution and returns y-list
    Args: lam - lambda value for Poisson distribution , type - float/int
          n - number of values(discrete) in x-axis to be taken , type-int
    Author: Lalith       
    """
    x = np.arange(n)
    f = lambda x: exp(-lam)*(lam**x)/np.math.factorial(x)
    y = []
    for i in range(len(x)):
        y.append(f(x[i]))
    return y

lam = float(input("Enter lambda:"))
n = int(input("Enter x axis limit:"))

x = np.arange(n)

plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)
plt.title('Poisson Distribution',fontsize=20)
plt.plot(x,poisson(lam,n),color='purple')
plt.show()
```

```python
# 4.Exponential Distribution

def exponential(lam,n):
    """
    Summary: This function will take lambda and n and creates a x - list and a corresponding y-list based on exponential distribution and returns y-list
    Args: lam - lambda value for exponential distribution , type - float/int
          n - x-axis limit , type-float/int
    Author: Lalith       
    """
    x = np.linspace(0,n,100)
    expr = lambda x: lam*exp(-lam*x)
    y = []
    for i in range(len(x)):
        y.append(expr(x[i]))
    return y

lam = float(input("Enter lambda(lambda>0):"))
n = int(input("Enter The x-axis limit:"))

x = np.linspace(0,n,100)

plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)
plt.title('Exponential Distribution',fontsize=20)
plt.plot(x,exponential(lam,n),color='tomato')
plt.show()
```

```python
# 5.Gamma Distribution

from sympy import *
import numpy as np
import matplotlib.pyplot as plt

def gamma_func(alpha):
    """
    Summary: This function will take alpha and returns gamma(alpha) 
    Args: alpha - alpha for gamma function , type - int/float
    Author: Lalith       
    """
    x = symbols('x')
    func = (x**(alpha-1))*exp(-x)
    y = integrate(func , (x,0,oo))
    return y

def function(x,alpha,lamda):
    """
    Summary: This function will take x , alpha, lambda values and returns the corresponding value based on the gamma distribution
    Args: x : x-axis value , type - float
          alpha: alpha for gamma distribution , type - float/int
          lamda: lamda fot gamma distribution , type - float/int
          
    Author: Lalith       
    """
    gal = gamma_func(alpha)
    f = lambda x: ((lamda**alpha)/gal)*(x**(alpha-1))*np.exp(-lamda*x)
    return(f(x))

a= float(input("Enter the starting x-value(x>0): "))
b= float(input("Enter the Ending x-value(x>0): "))
alpha = float(input("Enter alpha(alpha>0): "))
lamda = float(input("Enter lambda(lambda>0): "))
        
x = np.linspace(a,b,100)
y = []
for i in range(len(x)):
    y.append(function(x[i],alpha,lamda))
    
plt.xlabel('x',fontsize=18)
plt.ylabel('Gamma(x,alpha,lambda)',fontsize=10)
plt.title('Gamma Distribution',fontsize=20)
plt.plot(x,y,color='blue')
plt.show()
```
