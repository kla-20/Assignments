1. Use `matplotlib.pyplot.plot` to produce a plot of the functions 
   $f(x) = e^{-x/10}sin(πx)$ and $g(x) = xe^{−x/3}$ over the interval [0, 10]. 
   Include labels for the x- and y-axes, and a legend explaining which line is which plot.
   Save the plot as a .jpg (“Jpeg”).


```python
#1
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


f = lambda x : np.exp(-x/10)*np.sin(np.pi*x)
g = lambda x : x*np.exp(-x/3)

x = np.linspace(0,10,100)

y1 = f(x)
y2 = g(x)

plt.figure(figsize=(10,5))
plt.style.use('seaborn')


plt.title('Plot: $f(x) = e^{-x/10}sin(πx)$', fontsize=18)
plt.xlabel('X', fontsize=18)
plt.ylabel('F(X)', fontsize=18) 
plt.grid()
plt.plot(x,y1,'r',label='F(X)')
plt.legend(loc=8, fontsize=18)
#plt.savefig('plot1.jpeg',dpi=300)
plt.show()

plt.style.use('seaborn-dark')
plt.figure(figsize=(10,5))
plt.title('Plot: $g(x) = xe^{−x/3}$' , fontsize=18)
plt.xlabel('X', fontsize=18)
plt.ylabel('G(X)',fontsize=18) 
plt.grid()
plt.plot(x,y2,'g',label='G(X)')
plt.legend(loc=8, fontsize=18)
#plt.savefig('plot2.jpeg',dpi=300)
plt.show()
```

4. Use `matplotlib.pyplot.plot` to produce the surface and contour plot of the function $f(X) = x_1^2x_2^2(1-x_1-x_2).$ Can you identify maxima and minima using cantour plots. For better visualization, make interactive suface plot.

```python
#4
fig= plt.figure(figsize = (12,10))
plt.subplot(1,2,1)
ax = plt.axes(projection='3d')

x = np.linspace(-10,10,100)
y = x.copy()

x1, x2 = np.meshgrid(x,y)
Z = (x1**2)*(x2**2)*(1-x1-x2)

plt.title('Plot: $f(x) = x1^2*x2^2*(1-x1-x2)$', fontsize=18)
surf = ax.plot_surface(x1,x2,Z, cmap= plt.cm.viridis)

# set axes label
ax.set_xlabel('x', labelpad=20, fontsize=18)
ax.set_ylabel('y', labelpad=20, fontsize=18)
ax.set_zlabel('z', labelpad=7, fontsize=18)
fig.colorbar(surf, shrink=0.5, aspect=8) # adding colorbar
plt.savefig('plot5.jpeg',dpi=300)
plt.show()

#contour plot
fig,ax=plt.subplots(1,1)
cp = ax.contour(x1, x2, Z,cmap='Reds')
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Contour Plot',fontsize=20)
ax.set_xlabel('x (cm)',fontsize=15)
ax.set_ylabel('y (cm)',fontsize=15)
plt.show()
```

5. Plot gamma dribution for the parameters $(\alpha,\lambda),\ \alpha = 1, 2, 9; \ \lambda = 0.5,1$

$\gamma(\alpha,\lambda,x) = ((\lambda^\alpha)/(\gamma(\alpha))*\int (x^\alpha-1)*(\exp^(-\lambda*x))  ,LL=0, UL =oo$

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

def gamma_func(alpha):
    """
    Summary: This function takes alpha value and returns the gamma(alpha) value
    Args: alpha - alpha value for gamma function , type - int/float(alpha>0)
    Author: Lalith
    """
    x = symbols('x')
    func = (x**(alpha-1))*exp(-x)
    y = integrate(func , (x,0,oo))
    return y

def function(x,alpha,lamda):
    """
    Summary: This function takes x,alpha,lambda values and calculates gamma distribution value and returns it
    Args: alpha - alpha for gamma distribution , type - int/float(alpha>0)
          x - x-value for gamma distribution , type - int/float(x>0)
          lamda - lamda for gamma distribution , type - int/float(lamda>0)
    Author: Lalith
    """
    gal = gamma_func(alpha)
    f = lambda x: ((lamda**alpha)/gal)*(x**(alpha-1))*np.exp(-lamda*x)
    return(f(x))

a= float(input("Enter the starting x-value(x>0): "))
b= float(input("Enter the Ending x-value(x>0): "))

        
x = np.linspace(a,b,100)
yal1lam1 = []
yal1lam2 = []

yal2lam1 = []
yal2lam2 = []

yal3lam1 = []
yal3lam2 = []

for i in range(len(x)):
    yal1lam1.append(function(x[i],1,.5))

for i in range(len(x)):
    yal1lam2.append(function(x[i],1,1))
    
for i in range(len(x)):
    yal2lam1.append(function(x[i],2,.5))
    
for i in range(len(x)):
    yal2lam2.append(function(x[i],2,1))
    
for i in range(len(x)):
    yal3lam1.append(function(x[i],9,.5))
    
for i in range(len(x)):
    yal3lam2.append(function(x[i],9,1))
    

# Plotting the distributions

plt.figure(figsize=(14,8))
plt.suptitle('Gamma Distribution with Different alpha and lambda',fontsize=23,color='darkgreen')
plt.subplot(2,3,1) # 2-rows, 3-colums and 1st plot
plt.plot(x,yal1lam1,label='alpha = 1 ,lambda = 0.5',color='brown')
plt.title('alpha = 1,lambda = 0.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=12)
plt.grid()

# 2-rows, 3-colums and 2st plot
plt.subplot(2,3,2)
plt.plot(x,yal2lam1,label='alpha = 2 ,lambda = 0.5',color='orange')
plt.title('alpha = 2,lambda = 0.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=10.5)
plt.grid()

# 2-rows, 3-colums and 3rd plot
plt.subplot(2,3,3)
plt.plot(x,yal3lam1,label='alpha = 9 ,lambda = 0.5',color='deeppink') 
plt.title('alpha = 9,lambda = 0.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=2, fontsize=12)
plt.grid()

# 2-rows, 3-colums and 4th plot
plt.subplot(2,3,4)
plt.plot(x,yal1lam2,label='alpha = 1 ,lambda = 1',color='darkviolet') 
plt.title('alpha = 1,lambda = 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=12)
plt.grid()

# 2-rows, 3-colums and 5th plot
plt.subplot(2,3,5)
plt.plot(x,yal2lam2,label='alpha = 2 ,lambda = 1',color='darkred') 
plt.title('alpha = 2,lambda = 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=12)
plt.grid()

# 2-rows, 3-colums and 6th plot
plt.subplot(2,3,6)
plt.plot(x,yal3lam2,label='alpha = 9 ,lambda = 1',color='olive') 
plt.title('alpha = 9,lambda = 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=2, fontsize=10)
plt.grid()


plt.tight_layout()
plt.show()
```

6. Plot `log-normal` distribution with parameters $(\mu,\sigma),\ \mu = 0,1,2;\ \sigma= 0.5, 1, 2$

$ f(x) = (1/x\sigma\root(2\pi))\exp((-1/2)*((ln(x)-\mu)/\sigma)^2 $

```python
import numpy as np
import matplotlib.pyplot as plt


def ln_function(x,mu,var):
    """
    Summary: This function takes x,mean,variance and calculates for the log-normal distribution and returns the value
    Args: x - x-value for the log-normal distribution, type-int/float(x>0)
          mu - mean for the log normal distribution , type-int/float
          var - variance for the log normal distribution , type - int/float(var>0)
    Author: Lalith
    """
    coeff = 1/(x*np.sqrt(var)*np.sqrt(2*np.pi))
    expo = ((np.log(x)-mu)/np.sqrt(var))**2
    f = lambda x: coeff*np.exp(-.5*expo)
    return(f(x))

mu= float(input("Enter the mean for Log-Normal Distribution: "))
var= float(input("Enter the variance for Log-Normal Distribution (var>0): "))
a= float(input("Enter the starting x-value(x>0): "))
b= float(input("Enter the Ending x-value(x>0): "))

        
x = np.linspace(a,b,100)
ym1v1 = []
ym1v2 = []
ym1v3 = []

ym2v1 = []
ym2v2 = []
ym2v3 = []

ym3v1 = []
ym3v2 = []
ym3v3 = []

for i in range(len(x)):
    ym1v1.append(ln_function(x[i],0,.5))

for i in range(len(x)):
    ym1v2.append(ln_function(x[i],0,1))
    
for i in range(len(x)):
    ym1v3.append(ln_function(x[i],0,2))    
    
for i in range(len(x)):
    ym2v1.append(ln_function(x[i],1,.5))
    
for i in range(len(x)):
    ym2v2.append(ln_function(x[i],1,1))

for i in range(len(x)):
    ym2v3.append(ln_function(x[i],1,2))    
    
for i in range(len(x)):
    ym3v1.append(ln_function(x[i],2,.5))
    
for i in range(len(x)):
    ym3v2.append(ln_function(x[i],2,1))
    
for i in range(len(x)):
    ym3v3.append(ln_function(x[i],2,2))    
    

# Plotting the distributions

plt.figure(figsize=(14,12))
plt.suptitle('Log-Normal Distribution with Different mean and variance',fontsize=23,color='deeppink')

plt.subplot(3,3,1) # 2-rows, 3-colums and 1st plot
plt.plot(x,ym1v1,label='mue = 0 ,lambda = 0.5',color='brown')
plt.title('mue = 0 ,lambda = 0.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=12)
plt.grid()


plt.subplot(3,3,2)
plt.plot(x,ym2v1,label='mue = 1 ,lambda = 0.5',color='orange') 
plt.title('mue = 1 ,lambda = 0.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=10.5)
plt.grid()


plt.subplot(3,3,3)
plt.plot(x,ym3v1,label='mue = 2 ,lambda = 0.5',color='deeppink') 
plt.title('mue = 2 ,lambda = 0.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=4, fontsize=12)
plt.grid()


plt.subplot(3,3,4)
plt.plot(x,ym1v2,label='mue = 0 ,lambda = 1',color='darkviolet') 
plt.title('mue = 0 ,lambda = 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=12)
plt.grid()


plt.subplot(3,3,5)
plt.plot(x,ym2v2,label='mue = 1 ,lambda = 1',color='darkred') 
plt.title('mue = 1 ,lambda = 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=12)
plt.grid()


plt.subplot(3,3,6)
plt.plot(x,ym3v2,label='mue = 2 ,lambda = 1',color='olive') 
plt.title('mue = 2 ,lambda = 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=4, fontsize=10)
plt.grid()

plt.subplot(3,3,7)
plt.plot(x,ym1v3,label='mue = 0 ,lambda = 2',color='seagreen') 
plt.title('mue = 0 ,lambda = 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=10)
plt.grid()


plt.subplot(3,3,8)
plt.plot(x,ym2v3,label='mue = 1 ,lambda = 2',color='blue') 
plt.title('mue = 1 ,lambda = 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=10)
plt.grid()


plt.subplot(3,3,9)
plt.plot(x,ym3v3,label='mue = 2 ,lambda = 2',color='firebrick') 
plt.title('mue = 2 ,lambda = 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=1, fontsize=10)
plt.grid()



plt.tight_layout()
plt.show()
```
